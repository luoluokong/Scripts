```python
import tkinter as tk

# body 
root = tk.Tk()

# title
root.title("Title")

# calculate
def calculate(*args):
    try:
        value = float(feet.get())
        meter.set(int(0.3048 * value * 10000.0 + 0.5)/10000.0)
    except ValueError:
        pass

# create frame
mainframe = tk.ttk.Frame(root, padding="3 3 12 12")
mainframe.grid(column=0, row=0, sticky=(N, W, E, S))
root.columnconfigure(0, weight=1)
root.rowconfigure(0, weight=1)

# entry
feet = StringVar()
feet_entry = tk.ttk.Entry(mainframe, width=7, textvariable=feet)
feet_entry.grid(column=2, row=1, sticky=(W,E))


# remaining
meters = StringVar()
tk.ttk.Label(mainframe, textvariable=meters).grid(column=2, row=2, sticky=(W, E))
tk.ttk.Button(mainframe, text="Calculate", command=calculate).grid(column=3,\
                                                                   row=3,sticky=W)
ttk.Label(mainframe, text="feet").grid(column=3, row=1, sticky=W)
ttk.Label(mainframe, text="is equivalent to").grid(column=1, row=2, sticky=E)
ttk.Label(mainframe, text="meters").grid(column=3, row=2, sticky=W)

# binding
for child in mainframe.winfo_children(): 
    child.grid_configure(padx=5, pady=5)
feet_entry.focus()
root.bind("<Return>", calculate)



root.mainloop()
```
